```
import numpy as np   
import pandas as pd
import io
```

```
from google.colab import files
uploaded = files.upload()
```

```
data = pd.read_csv(io.BytesIO(uploaded['acquisitions.csv']))
```

```
print(data.head())
print(data.shape)
```

```
data_missing = data.isna()
print(data_missing.head())
```

```
print(data_missing.dtypes)
```

```
data_num_missing = data_missing.sum()
print(data_num_missing)
```

```
data.isna().mean().round(4) * 100
```

```
column_names = data.columns
for i in column_names:
  print('{} is unique: {}'.format(i, data[i].is_unique))
```

```
clean1=data
clean1.acquirer_category_list.fillna(clean1.company_category_list,inplace=True)
cleaned_data = clean1.dropna(how='any', subset=['acquirer_category_list'])
cleaned_data.isna().mean().round(4) * 100
```

```
cleaned_data = cleaned_data.dropna(how='any', subset=['company_state_code','company_category_list','company_region'])
cleaned_data.isna().mean().round(4) * 100
```

```
del cleaned_data['price_amount']
del cleaned_data['acquirer_state_code']
del cleaned_data['company_state_code']
```

```
cleaned_data.isna().mean().round(4) * 100
```

```
cleaned_data.acquirer_country_code.fillna(cleaned_data.company_country_code,inplace=True)
cleaned_data.acquirer_region.fillna(cleaned_data.company_region,inplace=True)
cleaned_data.acquirer_city.fillna(cleaned_data.company_city,inplace=True)
cleaned_data = cleaned_data.dropna(how='any', subset=['acquired_at'])
cleaned_data.price_currency_code.fillna('USD',inplace=True)
print(cleaned_data.isna().mean().round(4) * 100)
print(cleaned_data.isna().sum())
```

```
print(cleaned_data.shape)
```

```
cleaned_data.to_csv(r'dc_aquasitions.csv')
```
