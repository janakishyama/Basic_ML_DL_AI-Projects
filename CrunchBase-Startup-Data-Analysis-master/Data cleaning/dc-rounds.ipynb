```
import numpy as np   
import pandas as pd
import io
```

```
from google.colab import files
uploaded = files.upload()
```

```
data = pd.read_csv(io.BytesIO(uploaded['rounds.csv']))
```

```
print(data.head())
print(data.shape)
```

```
data_missing = data.isna()
print(data_missing.head())
```

```
print(data_missing.dtypes)
```

```
data_num_missing = data_missing.sum()
print(data_num_missing)
```

```
data.isna().mean().round(4) * 100
```

```
del data['funding_round_code']
del data['company_city']
del data['company_region']
```

```
del data['company_state_code']
```

```
clean1=data
clean1=clean1.dropna(how='any', subset=['company_category_list','company_name'])
clean1.company_country_code.fillna('USA',inplace=True)
median = clean1['raised_amount_usd'].median()
clean1.raised_amount_usd.fillna(median,inplace=True)
print(clean1.isna().mean().round(4) * 100)
print(clean1.isna().sum())
```
